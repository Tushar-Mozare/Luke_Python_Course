```python
import pandas as pd
from datasets import load_dataset
import numpy as np
import matplotlib.pyplot as plt

data = load_dataset("lukebarousse/data_jobs")
df = data['train'].to_pandas()

df.drop_duplicates(inplace = True)
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)

type(df.job_skills[1])
df.info()
```

```python
# Now I want to plot a pie chart of all boolean columns 
```

```python
dict = {'job_work_from_home' : 'Job Work From Home',
       'job_no_degree_mention' : 'Job No Degree Mention', 
       'job_health_insurance' : "Job Health Insurance"}

fig, ax = plt.subplots(1,3)

for i, (data, title) in enumerate(dict.items()) :
    df[data].value_counts().plot(kind = "pie", ax = ax[i], startangle = 90, autopct = "%1.1f%%")
    ax[i].set_title(title)
    ax[i].set_ylabel("")
    
fig.tight_layout()
```
