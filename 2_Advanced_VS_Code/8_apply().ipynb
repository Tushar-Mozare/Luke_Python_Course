```python
# In this module, we are going to learn how to use apply() function of pandas
```

```python
from datasets import load_dataset 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = load_dataset("lukebarousse/data_jobs")

df = data["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df.job_posted_date)

df.info()
```

```python
# Say I want to create a new column named salary_year_avg_modified where it get incremented by 5%
```

```python
df_og = df.copy()
```

```python
def salary_increment(sal) :
    return sal*1.05

df["salary_year_avg_modified"] = df.salary_year_avg.apply(salary_increment)

#OR df["salary_year_avg_modified"] = df.salary_year_avg.apply(lambda sal : sal*1.05)
df.info()
```

```python
df.dropna(subset = "salary_year_avg").head()
```

```python
# SO here apply() fuction will take the salary_year_avg series and then pass each item in the salary_increment function
# which get updated in salary_year_avg_modified series.
```

```python
# Syntax for df.apply() function :
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html
```

```python
# Now say I want to increase salary for suniors by 8% and for remainings by 3%
# For that I will apply apply() function on row so that I will get a series representong a row where 
# each item of that row can be called by their column name.
# Thus, we can apply conditions on items of one column to modify items of another column. 
```

```python
df1 = df_og.copy()

df1.info()
```

```python
def salary_increment_v2(row) :
    if "Senior" in row["job_title_short"] :
        return row["salary_year_avg"]*1.08
    else :
        return row["salary_year_avg"]*1.03
    
df1["salary_year_avg_modified"] = df1.apply(salary_increment_v2, axis=1)

# OR df1["salary_year_avg_modified"] = df1.apply(lambda row : row["salary_year_avg"]*1.08 if "Senior" in row["job_title_short"] else row["salary_year_avg"]*1.03, axis=1)
df1.info()
```
