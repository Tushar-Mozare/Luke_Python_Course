```python
job_skill = ["Python", "SQL",["SQL","Tableu"]]
```

```python
# There are 2 ways to syntax class method
```

```python
# 1st
list.append(job_skill, "R") # Here class_name.method_or_attribite(). Point to be noted first argument in the method_or_attribite() should be
                            # the variable on which you want to apply method or attribute.

print(job_skill)
```

```python
# 2nd
job_skill.remove("R") # Here variable.method_or_attribite(). Point to be noted Python identifies class from the variable data type.
                      # As variable is already define no need to define it again in method or attribute.

print(job_skill)
```

```python
# Point to be noted 1st syntax method is general method and can be use on any class
# Whereas 2nd syntax method is special just for predefined classes
```

```python
# From the example itself you know the use of append() and remove() methods.
```
