```python
# Prac 1
# Create a code for giving me details of candidate whether he has the skill but not enough exp
# or have enough exp but not in required skill
# or have enough exp in the required skill; Hence Selected!
# Use if elif statement to create the code.

# Heres the input you required.
c1_job_title = "Python"
c1_exp = 3

c2_job_title = "Java"
c2_exp = 6

c3_job_title = "Python"
c3_exp = 5

# Job Title is Python and min exp required is 5
```

```python
def selection(c_job_title, job_title, c_exp, exp):
  if c_job_title == job_title and c_exp < exp:
    return "Candidate has the skill but not enough exp."
  elif c_job_title != job_title and c_exp >= exp:
   return "Candidate has enough exp but not in required skill."
  elif c_job_title == job_title and c_exp >= exp:
    return "Candidate has enough exp in the required skill; Hence Selected!"
```

```python
job_title = "Python"
exp = 5
```

```python
c1_vardict = selection(c1_job_title, job_title, c1_exp, exp)
c2_vardict = selection(c2_job_title, job_title, c2_exp, exp)
c3_vardict = selection(c3_job_title, job_title, c3_exp, exp)
```

```python
final_verdict = f"Candidate 1: {c1_vardict}\nCandidate 2: {c2_vardict}\nCandidate 3: {c3_vardict}"
```
