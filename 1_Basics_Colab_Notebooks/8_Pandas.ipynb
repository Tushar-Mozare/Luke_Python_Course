```python
# Say we want to extract last five entries of the longitude column
```

```python
df.longitude[-5:] # OR df.longitude.tail()
```

```python
# In slicing :
# Format : [f : l]
# Always remember that the number we will put in l will be skipped
# So in last example if you code df.longitude[-5:-1] then it will skip last entry
# To solve that python provide some interpretation
# IF l is blank : It represents to have the slice till the last entry
#    f is blank : It represents to have the slice from the first entry
# Thats why  df.longitude[-5:]
```
