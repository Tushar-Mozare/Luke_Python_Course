```python
# Now say we want to replace some str characters with new ones so how we are gonna do that.
```

```python
job_title = "Doto Analyst"
```

```python
# See "Data Analyst" is mispelled as "Doto Analyst". So let act as your girlfriend : "I CAN FIX HIM"
```

```python
job_title = job_title.replace("o","a",2)
print(job_title)
```

```python
# Observe that in str.replace("old","new",count=int), we first enter the word or character to be replaced then we enter the word or character
# with whom you are gonna to replace and then we specifiy to how many occarance it should be replace
# IOC we enter 2, it means first two "o"s will be going to replaced with "a"s
# whatif we want to replace all occarances then we will put "-1" OR leave it blank as by default it is set to -1
```

```python
job_title = job_title.replace("a","o")
print(job_title)
```

```python
# OR
```

```python
job_title = job_title.replace("o","a",-1)
print(job_title)
```

```python
# Let split the String
```

```python
# Syntax : str.split(sep=None,maxsplit=-1)
# "sep" specify the character or grp of characters from which we want to split str
# "maxsplit" how many occurance of "sep" to be consider for spliting the string ; -1 set as default and representing all occurance
# Lemme Clarify with example
```

```python
sad_dr = "An Apple a day keeps doctor away"
```

```python
sad_dr.split(sep=" ",maxsplit=4) # OR : sad_dr.split(" ",4)
```

```python
sad_dr.split(sep=" ",maxsplit=-1) # OR : sad_dr.split(" ",-1)
```

```python
# Lets talk about string formatting
# There are 3 methods to do it
# 1. format str
# 2. f-string
# 3. % style format str
```

```python
name = "Tushar Mozare"
age = int(21)
nat = "INDIA"
role = "Data Analyst Jr."
```

```python
# 1. format str
```

```python
e1 = "Name : {}, Age : {}, Nationality : {}, Role : {}".format(name,age,nat,role)
print(e1)
```

```python
# 2. f-string
```

```python
e2 = f"Name : {name}, Age : {age}, Nationality : {nat}, Role : {role}"
print(e2)
```

```python
# 3. % style format str
```

```python
e3 = "Name : %s, Age : %s, Nationality : %s, Role : %s"%(name,age,nat,role)
print(e3)
```
